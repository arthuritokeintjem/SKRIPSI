**Baca Data**

```python
import pandas as pd

df = pd.read_csv('dataset_raw.csv')  # Ganti dengan nama file Anda
```

**Konversi Kolom Tanggal ke Datatime**

```python
df['tanggal'] = pd.to_datetime(df['tanggal'])
```

**Tambahkan Kolom Hari**

```python
df['hari'] = df['tanggal'].dt.dayofweek  # 0=Senin, ..., 5=Sabtu, 6=Minggu
```

**Filter Data Weekday Saja**

```python
df_weekday = df[(df['hari'] < 5) & (df['harga'] > 0)]  # Hanya Senin-Jumat dan harga > 0
```

**Hapus Kolom Hari**

```python
df_weekday['nama_hari'] = df_weekday['tanggal'].dt.day_name(locale='id_ID')  # Nama hari dalam Bahasa Indonesia

df_weekday = df_weekday.drop(columns=['hari'])
```

**Simpan Data Baru**

```python
df_weekday.to_csv('dataset.csv', index=False)
```
